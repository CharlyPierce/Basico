```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

#Compresi√≥n sin eigenvalores ni eigen vectores


```
#compresion sin eigenvalor ni eigenvectores
image1=io.imread("3.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
R=30
#from sklearn.preprocessing import StandardScaler
#sc=StandardScaler()
#image=sc.fit_transform(image1)
image=image1
# en el rango 0-255. Por eso la convertimos a flotante y en el rango 0-1
print("- Dimensiones de la imagen:")
print(image.shape)
an=[]
for i in range(R):
  an.append(image[:,np.random.randint(0,image.shape[1])])
AT=np.array(an)
A=AT.T
AI=np.linalg.inv(np.dot(AT,A))
M_P=np.dot(np.dot(A,AI),AT)
an1=[]
for i in range(image.shape[1]):
  an1.append(np.dot(M_P,image[:,i]))
N_image=np.array(an1).T
print(plt.imshow(image))#,vmin=0,vmax=1))
plt.figure()
print(plt.imshow(N_image,vmin=0,vmax=1))
```

```
#compresion sin eigenvalor ni eigenvectores
ran=30
image=io.imread("11.jpg")/255.0 # imread lee las imagenes con los pixeles codificados como enteros 
A=image
#Matrices U sigma V calculadas
S=np.linalg.svd(A)
U=S[0]
sigma=np.zeros((S[0].shape[1],S[2].shape[0]))
for i in range(len(S[1])):
  if(sigma.shape[1]>1):
    sigma[i][i]=S[1][i]
  else:
    sigma[i]=S[1][i]
V=S[2]
#
N_I=np.dot(np.dot(U[:,:ran],sigma[:ran,:ran]),V[:ran,:])
m_r = np.linalg.matrix_rank(A)#da un rango bueno k
#SVD #comprobacion
A_Q=np.dot(np.dot(U,sigma),V)#multiplicacion U S V 

print("- Dimensiones de la imagen:")
print(image.shape)
print(plt.imshow(image))#,vmin=0,vmax=1))
plt.figure()
plt.imshow(N_I)
```
